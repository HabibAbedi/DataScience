```python
from bidi.algorithm import get_display
import arabic_reshaper
```

<div class="alert alert-success" role="alert" dir="rtl">
    برای نوشتن متن فارسی(عربی) ابتدا با استفاده از 
    <span class="label label-primary">()arabic_reshaper.reshape</span>،
    متن را تغییر شکل میدهیم. سپس با تابع
    <span class="label label-primary">()get_display</span>
    متن را نمایش میدهیم.
    برای استفاده راحتتر تابع زیر را تعریف میکنیم.
</div>

```python
def persian(x):
    reshaped = arabic_reshaper.reshape(x)
    persian = get_display(reshaped)
    return persian
```

<div class="alert alert-success" role="alert" dir="rtl">
برای تغییر اندازه و نوع و رنگ فونت به صورت زیر عمل میکنیم.
</div>

```python
font_title = {'family': 'B Farnaz',
              'color':  'red',
              'weight': 'normal',
              'size': 30,
              }
font_labels = {'family': 'B Nazanin',
                'color':  'black',
                'weight': 'normal',
                'size': 20,
                }
 
plt.plot(X,y)
plt.xlabel(make_farsi_text(xlabel),fontdict = font_labels)
plt.ylabel(make_farsi_text(ylabel),fontdict = font_labels)
plt.title(make_farsi_text(title) ,fontdict = font_title)
ax = plt.gca()
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(14)
    tick.label.set_rotation('0')
    tick.label.set_fontname('B Lotus')
    
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(14)
    tick.label.set_rotation('0')
    tick.label.set_fontname('B Lotus')
    
plt.show()
```
